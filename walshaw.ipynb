```python
import networkx as nx
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import time
import altair as alt
import math
```

```python
basepath = "data/walshaw/"

def read_graph(collection, index):
    
    graph_file = os.listdir(basepath + collection)[index]
    f = open(basepath + collection + "/" + graph_file, "r")
    lines = f.readlines()

    graph = nx.Graph()
    
    for i in range(int(lines[0].strip().split(" ")[0])):
        graph.add_node(i)

    for i in range(1, len(lines)):
        if lines[i].strip() == "": continue
        line = lines[i].strip().split(" ")
        for j in range(len(line)):
            if int(line[j].strip()) > i : graph.add_edge(i, line[j])

    return graph

graph = read_graph("graphs", 10)
```

```python
arrstats = []
for i in range(len(os.listdir(basepath + "graphs"))):
    print(i)
    graph = read_graph("graphs", i)
    # print(len(graph.nodes()))
    arrstats.append({
        'nodes': len(graph.nodes()),
        'edges': len(graph.edges()),
        'degree': graph.degree()
    })
```

```python
numnodedict = {}
for g in arrstats:
    if g['nodes'] not in numnodedict:
        numnodedict[g['nodes']] = 1
    else: numnodedict[g['nodes']] += 1

source = pd.DataFrame({
    'node count': list(numnodedict.keys()),
    'occurrences': list(numnodedict.values())
})

alt.Chart(source, title="Distribution of node counts across graphs in Walshaw collection").mark_bar().encode(
    x='node count',
    y='occurrences'
)
```

```python
numnodedict = {}
for g in arrstats:
    if g['edges'] not in numnodedict:
        numnodedict[g['edges']] = 1
    else: numnodedict[g['edges']] += 1

source = pd.DataFrame({
    'edge count': list(numnodedict.keys()),
    'occurrences': list(numnodedict.values())
})

alt.Chart(source, title="Distribution of edge counts across graphs in Walshaw collection").mark_bar().encode(
    x='edge count',
    y='occurrences'
)
```

```python
avgdegrees = {}
for g in arrstats:
    curavg = 0
    for deg in g['degree']:
        curavg += deg[1]
    curavg /= len(g['degree'])
    curavg = round(curavg*10)/10
    
    if curavg not in avgdegrees:
        avgdegrees[curavg] = 1
    else: avgdegrees[curavg] += 1

source = pd.DataFrame({
    'avg degree in graph': list(avgdegrees.keys()),
    'occurrences': list(avgdegrees.values())
})

alt.Chart(source, title="Distribution of avg degree across graphs in Walshaw collection").mark_bar().encode(
    x='avg degree in graph',
    y='occurrences'
)
```

```python
avgdegrees = {}
for g in arrstats:
    curmax = 0
    for deg in g['degree']:
        if deg[1] > curmax:
            curmax = deg[1]
    if curmax not in avgdegrees:
        avgdegrees[curmax] = 1
    else: avgdegrees[curmax] += 1

source = pd.DataFrame({
    'max degree in graph': list(avgdegrees.keys()),
    'occurrences': list(avgdegrees.values())
})

alt.Chart(source, title="Distribution of max degree across graphs in Walshaw collection").mark_bar().encode(
    x='max degree in graph',
    y='occurrences'
)
```
