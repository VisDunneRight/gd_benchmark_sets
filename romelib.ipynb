```python
import networkx as nx
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import time
import altair as alt
import math
```

```python
def plot_degrees(graph):
    degdict = {}
    for deg in graph.degree():
        if deg[1] not in degdict:
            degdict[deg[1]] = 1
        else: degdict[deg[1]] += 1

    print(list(degdict.keys()))
    print(list(degdict.values()))

    source = pd.DataFrame({
        'a': list(degdict.keys()),
        'b': list(degdict.values())
    })

    return alt.Chart(source).mark_bar().encode(
        x='a',
        y='b'
    )
```

```python
basepath = "data/rome/"

print("number of graphs: ", len(os.listdir(basepath)) - 1)
arrstats = []

for i, filename in enumerate(os.listdir(basepath)):
    if filename == "Graph.log": continue
    graph = nx.read_graphml(basepath + filename)
    arrstats.append({
        'nodes': len(graph.nodes()),
        'edges': len(graph.edges()),
        'degree': graph.degree()
    })
```

```python
numnodedict = {}
for g in arrstats:
    if g['nodes'] not in numnodedict:
        numnodedict[g['nodes']] = 1
    else: numnodedict[g['nodes']] += 1

source = pd.DataFrame({
    'node count': list(numnodedict.keys()),
    'occurrences': list(numnodedict.values())
})

alt.Chart(source, title="Distribution of node counts across graphs in Rome-Lib").mark_bar().encode(
    x='node count',
    y='occurrences'
)
```

```python
numnodedict = {}
for g in arrstats:
    if g['edges'] not in numnodedict:
        numnodedict[g['edges']] = 1
    else: numnodedict[g['edges']] += 1

source = pd.DataFrame({
    'edge count': list(numnodedict.keys()),
    'occurrences': list(numnodedict.values())
})

alt.Chart(source, title="Distribution of edge counts across graphs in Rome-Lib").mark_bar().encode(
    x='edge count',
    y='occurrences'
)
```

```python
avgdegrees = {}
for g in arrstats:
    curavg = 0
    for deg in g['degree']:
        curavg += deg[1]
    curavg /= len(g['degree'])
    curavg = round(curavg*10)/10
    
    if curavg not in avgdegrees:
        avgdegrees[curavg] = 1
    else: avgdegrees[curavg] += 1

source = pd.DataFrame({
    'avg degree in graph': list(avgdegrees.keys()),
    'occurrences': list(avgdegrees.values())
})

alt.Chart(source, title="Distribution of avg degree across graphs in Rome-Lib").mark_bar().encode(
    x='avg degree in graph',
    y='occurrences'
)
```

```python
avgdegrees = {}
for g in arrstats:
    curmax = 0
    for deg in g['degree']:
        if deg[1] > curmax:
            curmax = deg[1]
    if curmax not in avgdegrees:
        avgdegrees[curmax] = 1
    else: avgdegrees[curmax] += 1

source = pd.DataFrame({
    'max degree in graph': list(avgdegrees.keys()),
    'occurrences': list(avgdegrees.values())
})

alt.Chart(source, title="Distribution of max degree across graphs in Rome-Lib").mark_bar().encode(
    x='max degree in graph',
    y='occurrences'
)
```
